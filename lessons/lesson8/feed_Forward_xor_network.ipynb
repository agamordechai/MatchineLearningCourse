```python
import numpy as np
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
X=np.array([[-1,-1],[-1,1],[1,-1],[1,1]])
```

```python
W1=np.array([[-5,-5],[5,5]])
B1=np.array([[5],[5]])
W2=np.array([[6,6]])
B2=-10

a0 = X.T
a1 = sigmoid(np.dot(W1,a0)+B1)
a2 = sigmoid(np.dot(W2,a1)+B2)
print('a0\n',a0)
print('a1\n',a1)
print('a2\n',a2)
```
