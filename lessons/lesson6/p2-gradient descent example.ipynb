```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
  return x**2 - 10 * x + 5
```

```python
def dx(x):
  return 2*x - 10
```

```python
alpha = 0.1
```

```python
x=0

print ("epoch\tx\tdx\tα·dx")
for i in range(5):
  print ("%d\t%.2f\t%.2f\t%.2f" % (i,x,dx(x),alpha * dx(x)))
  x = x - alpha * dx(x)
```

```python
for alpha in [0.01,0.03,0.1,0.3]:
  x=0
  x_history=[x]

  for i in range(40):
    x = x - alpha * dx(x)
    x_history.append(x)
  
  plt.plot(x_history,label=str(alpha))
plt.legend()
plt.grid()
plt.xlabel('epochs');
plt.ylabel('x');
```
